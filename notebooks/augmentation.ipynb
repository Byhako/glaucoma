```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import matplotlib.pyplot as plt
import os
```

## Creamos directorios y lista de imagenes

```python
folder = '/content/drive/MyDrive/Glaucoma'
folder_target_name = 'data/oringa/augmentation/original'
folder_origin_name = 'data/oringa/original'

os.makedirs(f"{folder}/{folder_target_name}/Train/negative", exist_ok=True)
os.makedirs(f"{folder}/{folder_target_name}/Train/positive", exist_ok=True)
os.makedirs(f"{folder}/{folder_target_name}/Test/negative", exist_ok=True)
os.makedirs(f"{folder}/{folder_target_name}/Test/positive", exist_ok=True)

test_positive = os.listdir(f"{folder}/{folder_origin_name}/Test/positive")
test_negative = os.listdir(f"{folder}/{folder_origin_name}/Test/negative")
train_positive = os.listdir(f"{folder}/{folder_origin_name}/Train/positive")
train_negative = os.listdir(f"{folder}/{folder_origin_name}/Train/negative")
```

## Función para generar imágenes

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    brightness_range=[0.5, 1.5]
)

def dataAugmentation(image_names, folder_name, n_images):
    """
    image_names: list of image names
    folder_name: name of the folder of every class
    n_images: number of images to be generated
    """
    origin_folder = f"{folder}/{folder_origin_name}/{folder_name}"
    target_folder = f"{folder}/{folder_target_name}/{folder_name}"
    total = len(image_names)
    j = 1

    for path in image_names:
        print(f"{j}/{total}")
        j += 1
        img = load_img(f"{origin_folder}/{path}")
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape)

        i = 0
        for batch in datagen.flow(x, batch_size=1, save_to_dir=f"{target_folder}", save_prefix='m', save_format='jpg'):
            i += 1
            if i == n_images:
                break
```

## Usamos función

```python
dataAugmentation(test_positive, 'Test/positive', 4)
```

```python
dataAugmentation(test_negative, 'Test/negative', 4)
```

```python
dataAugmentation(train_negative, 'Train/negative', 4)
```

```python
dataAugmentation(train_positive, 'Train/positive', 4)
```
