```python
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import os
```

## Creamos directorios y lista de imagenes

```python
folder = '/content/drive/MyDrive/Glaucoma'
folder_target_name = 'data/oringa/inversion/augmentation'
folder_origin_name = 'data/oringa/augmentation/original'

os.makedirs(f"{folder}/{folder_target_name}/Train/negative", exist_ok=True)
os.makedirs(f"{folder}/{folder_target_name}/Train/positive", exist_ok=True)
os.makedirs(f"{folder}/{folder_target_name}/Test/negative", exist_ok=True)
os.makedirs(f"{folder}/{folder_target_name}/Test/positive", exist_ok=True)

test_positive = os.listdir(f"{folder}/{folder_origin_name}/Test/positive")
test_negative = os.listdir(f"{folder}/{folder_origin_name}/Test/negative")
train_positive = os.listdir(f"{folder}/{folder_origin_name}/Train/positive")
train_negative = os.listdir(f"{folder}/{folder_origin_name}/Train/negative")
```

## Funci√≥n para invertir color

```python
def invertir(image_names, folder_name):
    origin_folder = f"{folder}/{folder_origin_name}/{folder_name}"
    target_folder = f"{folder}/{folder_target_name}/{folder_name}"

    for path in image_names:
        imagen = Image.open(f'{origin_folder}/{path}')
        imagen_invertida = ImageOps.invert(imagen)
        imagen_invertida.save(f'{target_folder}/{path}')
```

```python
invertir(test_positive, 'Test/positive')
invertir(test_negative, 'Test/negative')
invertir(train_positive, 'Train/positive')
invertir(train_negative, 'Train/negative')
```

```python
folder = '/content/drive/MyDrive/Glaucoma'
folder_target_name = 'data/oringa/inversion/augmentation'
folder_origin_name = 'data/oringa/augmentation/original'

original = Image.open(f'{folder}/data/oringa/original/Test/positive/598.jpg')
invertida = Image.open(f'{folder}/data/oringa/inversion/original/Test/positive/598.jpg')

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Mostrar la primera imagen
ax[0].imshow(original)
ax[0].set_title('Original')
ax[0].axis('off')  # Ocultar ejes

# Mostrar la segunda imagen
ax[1].imshow(invertida)
ax[1].set_title('Invertida')
ax[1].axis('off')  # Ocultar ejes

# Mostrar la figura
plt.show()
```
